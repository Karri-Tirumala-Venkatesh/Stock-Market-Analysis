```python
import yfinance as yf
import pandas as pd

# Define stock symbol for NSE
ticker = "RELIANCE.NS"

# Fetch minute-wise data for the last 1 day
df = yf.download(ticker, period="1d", interval="1m")

# Check if the index is already timezone-aware
if not df.index.tz:
    # If not, localize to UTC
    df.index = df.index.tz_localize('UTC')

# Convert the index to Indian Standard Time (IST)
df.index = df.index.tz_convert('Asia/Kolkata')

df
```

```python
# Convert df to a csv file which can be downloaded
from google.colab import files
df.to_csv('reliance.csv')
files.download('reliance.csv')
```
